# Работаем с жанрами

```python
movies['genres_split'] = movies.genres.str.split('|')
```

```python
movies['genres_cnt'] = movies.apply(lambda r: len(r['genres_split']), axis=1)
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
```

```python
movies['genres_space'] = movies.apply(lambda r: ' '.join(r['genres_split']), axis=1)
```

```python
movies.head()
```

```python
cnt_vec_gen = CountVectorizer()
```

```python
processed_gen = cnt_vec_gen.fit_transform(movies['genres_space'])
```

```python
tfidf_gen = TfidfTransformer()
```

```python
tfidf_dense_gen = tfidf_gen.fit_transform(processed_gen).todense()
```

```python
cnt_vec_gen.vocabulary_
```

```python
columns_gen = [None for i in range(len(cnt_vec_gen.vocabulary_))]
for k in cnt_vec_gen.vocabulary_:
    columns_gen[cnt_vec_gen.vocabulary_[k]] = k
```

```python
df_tfidf_gen = pd.DataFrame(tfidf_dense_gen, columns=columns_gen)
```

```python
df_tfidf_gen.head()
```

```python
movies_with_tfidf = pd.concat((movies, df_tfidf_gen), axis=1)
```

```python
movies_with_tfidf.head()
```

```python
import re
movies_with_tfidf['year'] = movies_with_tfidf.apply(lambda r: int(re.findall(r'\(([0-9]{4})\)', r['title'])[0]) if len(re.findall(r'\(([0-9]{4})\)', r['title'])) else None, axis=1)
```

```python
movies_with_tfidf['year'] = movies_with_tfidf['year'].fillna(movies_with_tfidf['year'].median())
```

```python
movies_with_tfidf.columns
```

```python
fc_gen = ['movieId', 'title', 'genres_cnt', 'action', 'adventure', 'animation', 'children',
       'comedy', 'crime', 'documentary', 'drama', 'fantasy', 'fi', 'film',
        'horror', 'imax', 'listed', 'musical', 'mystery', 'no',
       'noir', 'romance', 'sci', 'thriller', 'war', 'western', 'year']
```

```python
movies_for_ds = movies_with_tfidf[fc_gen]
```

# Работаем с оценками

```python
ratings = pd.read_csv('/Users/mamolchanova/Documents/Обучение/ML/15 Занятие - рекомендательные системы/ratings.csv')
```

```python
ratings.head()
```

```python
rating_movies = ratings.groupby("movieId")
rating_movies_mean = rating_movies.agg({"rating": "mean"}).rename({'rating': 'rating_movies_mean'}, axis=1)
rating_movies_mean.head()
```

```python
rating_movies_std = rating_movies.agg({"rating": "std"}).rename({'rating': 'rating_movies_std'}, axis=1)
rating_movies_std.head()
```

```python
rating_movies_mean = rating_movies_mean.reset_index()
```

```python
rating_movies_std = rating_movies_std.reset_index()
```

```python
rating_users = ratings.groupby("userId")
rating_users_mean = rating_users.agg({"rating": "mean"}).rename({'rating': 'rating_users_mean'}, axis=1).reset_index()
rating_users_std = rating_users.agg({"rating": "std"}).rename({'rating': 'rating_users_std'}, axis=1).reset_index()
```

# Строим предсказательную модель

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
TARGET_USER = 474
```

```python
df_for_user = movies_final[movies_final['userId']==TARGET_USER]
```

```python
# у некоторых фильмов, которым была поставлена только одна оценка, дисперсия=NaN. И пустое среднее
# чтобы не было ошибки, заполняю х 0

for i in df_for_user.columns:
    df_for_user[i]=df_for_user[i].fillna(0)
```

```python
y = df_for_user['rating']
```

```python
X = df_for_user.drop(['tag', 'tag_space', 'rating', 'userId', 'movieId', 'timestamp', 'title', 'genres', 'genres_split', 'genres_space'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
models = [LinearRegression, Lasso, Ridge, RandomForestRegressor, SVR]
```

```python
for m in models:
    model = m()
    model.fit(X_train, y_train)
    print("{}. r2_train: {:.4f}, r2_test: {:.4f}, mae_train: {:.4f}, mae_test: {:.4f}".format(
        m.__name__, model.score(X_train ,y_train), model.score(X_test, y_test),
        mean_absolute_error(model.predict(X_train), y_train),
        mean_absolute_error(model.predict(X_test), y_test)
    ))
```

# Улучшаю качество

Мне не нравится качество моделей. Особенно, как ведет себя линейная регрессия. Есть гипотеза, что если уменьшить размерность, результаты станут лучше.

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=500, n_iter=7, random_state=42)
```

```python
svd.fit(X_train)
```

```python
print(svd.explained_variance_ratio_.sum())
```

```python
X_train_svd = svd.transform(X_train)
```

```python
X_test_svd=svd.transform(X_test)
```

```python
for m in models:
    model = m()
    model.fit(X_train_svd, y_train)
    print("{}. r2_train: {:.4f}, r2_test: {:.4f}, mae_train: {:.4f}, mae_test: {:.4f}".format(
        m.__name__, model.score(X_train_svd, y_train), model.score(X_test_svd, y_test),
        mean_absolute_error(model.predict(X_train_svd), y_train),
        mean_absolute_error(model.predict(X_test_svd), y_test)
    ))
```

В данном случае, линейная регрессия показывает уже адекватные значения. У других моделей показатели тоже немного поменялись (хотя это может быть связано с изменением разбиения.
